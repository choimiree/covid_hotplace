```python
def yd_cr(yd):
    yd.drop(['시간(1시간단위)','성별', '시'], axis=1, inplace=True)
    yd.columns = ["날짜","연령대","구","유동인구수"]
    yd_return = yd.groupby(['날짜', '구', "연령대"]).sum()
    yd2 = yd_return.reset_index()
    return yd2
```

```python
df_yd = pd.DataFrame(columns = ["날짜","연령대","구","유동인구수"])

for k in range (2,9):
    data = pd.read_csv(f"유동인구/Floating_Population_200{k}.csv", encoding = 'utf-8')
    yd = yd_cr(data)
    df_yd = df_yd.append(yd)



```
