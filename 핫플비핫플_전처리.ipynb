```python
def yd_cr(yd):
    yd.drop(['시간(1시간단위)','성별', '시'], axis=1, inplace=True)
    yd.columns = ["날짜","연령대","구","유동인구수"]
    yd_return = yd.groupby(['날짜', '구', "연령대"]).sum()
    yd2 = yd_return.reset_index()
    return yd2
```

```python
df_yd = pd.DataFrame(columns = ["날짜","연령대","구","유동인구수"])

for k in range (2,9):
    data = pd.read_csv(f"유동인구/Floating_Population_200{k}.csv", encoding = 'utf-8')
    yd = yd_cr(data)
    df_yd = df_yd.append(yd)



```

# 연령별총합

```python
#total 구하기. 피봇테이블에는 인덱스로 봐야하기 때문에! 따옴표 안쓴다!!! 
#yd_month['총합'] = yd_month[20] + yd_month[30] + yd_month[40] + yd_month[50] + yd_month[60] + yd_month[70]

#근데 우리는 아까 위에서 피봇테이블->일반테이블해줬기때문에 컬럼명을 써야한다!

#일단은 모든 컬럼 이름으로 더해줬는데, 인덱싱해서 가져오고 싶어요. 컬럼 인덱싱은 어떻게 하는 걸가요?

yd_new['날짜총합'] = yd_new[['20', '30', '40', '50', '60', '70']].sum(axis=1)

```
